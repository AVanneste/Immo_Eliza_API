```python
import pandas as pd
import json

# df = pd.read_csv("cleaned_data.csv")
# df = df[['id', 'type of property', 'Sub-type', 'City', 'Province', 'Region', 'post code', 'Price', 'Median price PC', 'Garden', 'Terrace','Living area','Bedrooms','Surface of the plot','Garden surface','Swimming pool','Open fire','Terrace surface','Number of frontages','Building condition', 'Kitchen type']]
df= pd.read_csv('coef.csv')
df.head()
```

```python
prop_type_list = ['APARTMENT','HOUSE','OTHERS']
# corr_list_of_list = []
corr_dict = {}

for i in prop_type_list:
    corr = df[df['type of property'] == i].corr()
    corr_list = dict(corr['Price'])
    # corr_list_of_list.append(corr_list)
    corr_dict[i] = corr_list


with open("correlations.json", "w") as outfile:
    json.dump(corr_dict, outfile, indent=4)

```

```python
national_median = df['Price'].median()

pc_coefs = dict(zip(df['post code'], df['Median price PC']/national_median))

with open("pc_coefs.json", "w") as outfile:
    json.dump(pc_coefs, outfile, indent=4)
```

```python
num_col_list = ['Living area','Bedrooms','Surface of the plot','Garden surface','Terrace surface','Number of frontages','Building condition', 'Kitchen type']
prop_type_list = ['APARTMENT','HOUSE','OTHERS']
median_dict = {}

for i in prop_type_list:
    dict1 = {}
    for j in num_col_list:
        prop_type_mean = df[(df['type of property']== i) & (df[j] != 0) ][j].median()
        # df.loc[(df['type of property'] == i ), j] = df[j]/prop_type_mean
        dict1[j] = prop_type_mean
    median_dict[i] = dict1

with open("medians.json", "w") as outfile:
    json.dump(median_dict, outfile, indent=4)
```
